```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import math
import os
import glob
```

```python
img_mandrill = cv.imread('/content/drive/MyDrive/mandrill.png')
img_rgb = cv.cvtColor(img_mandrill,cv.COLOR_BGR2RGB) # konversi channel BGR -> RGB
plt.imshow(img_rgb) 

height, width, depth = np.shape(img_mandrill)
names = np.arange(256)

blue = [0]*256
green = [0]*256
red = [0]*256

for y in range(0,height):
  for x in range(0,width):
    blue[img_mandrill[y][x][0]] += 1
    green[img_mandrill[y][x][1]] += 1
    red[img_mandrill[y][x][2]] += 1

names = np.arange(256)
fig, axs = plt.subplots(1, 3, figsize=[20,5], sharex=True, sharey=True)
fig.suptitle('Histogram RGB plot')
fig.text(0.05, 0.5, 'Jumlah Kemunculan', va='center', rotation='vertical')
fig.text(0.5, 0.04, 'Intensitas Warna', ha='center')
axs[0].bar(names, red, color='red')
axs[1].bar(names, green, color='green')
axs[2].bar(names, blue, color='blue')
```
